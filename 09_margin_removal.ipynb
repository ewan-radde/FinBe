```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/E2_24-25.csv")

odds = df[["HomeTeam","AwayTeam","FTHG","FTAG","PSCH","PSCD","PSCA"]].copy()
odds = odds.dropna(subset=["PSCH","PSCD","PSCA"]).reset_index(drop=True)

# raw implied probabilities (inflated by margin)
odds["imp_home"] = 1 / odds["PSCH"]
odds["imp_draw"] = 1 / odds["PSCD"]
odds["imp_away"] = 1 / odds["PSCA"]
odds["overround"] = odds["imp_home"] + odds["imp_draw"] + odds["imp_away"]

print(f"Matches: {len(odds)}")
print(f"Average overround: {odds['overround'].mean():.4f}")
odds[["PSCH","PSCD","PSCA","imp_home","imp_draw","imp_away","overround"]].head()
```
