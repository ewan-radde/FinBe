```python
import pandas as pd
import numpy as np

# load the most recent complete season (richest odds coverage)
df = pd.read_csv("data/E2_25-26.csv")

print(f"Total columns: {len(df.columns)}")
print(f"\nAll columns:\n{df.columns.tolist()}")

# specifically look for the bookmaker odds columns we care about
# H/D/A suffixes = home/draw/away odds for each bookmaker
odds_cols = [c for c in df.columns if c.endswith(("H","D","A")) and len(c) <= 6]
print(f"\nLikely odds columns:\n{odds_cols}")
```

```python
# Pinnacle closing odds
odds = df[["HomeTeam", "AwayTeam", "FTHG", "FTAG", "PSCH", "PSCD", "PSCA"]].copy()

# drop any matches missing Pinnacle odds (can't compute those)
odds = odds.dropna(subset=["PSCH", "PSCD", "PSCA"]).reset_index(drop=True)

# raw implied probability = 1 / decimal odds
odds["imp_home"] = 1 / odds["PSCH"]
odds["imp_draw"] = 1 / odds["PSCD"]
odds["imp_away"] = 1 / odds["PSCA"]

# overround = sum of the three raw implied probabilities
odds["overround"] = odds["imp_home"] + odds["imp_draw"] + odds["imp_away"]

print(f"Matches with Pinnacle closing odds: {len(odds)}")
print(f"\nAverage overround: {odds['overround'].mean():.4f}")
print(f"Min overround: {odds['overround'].min():.4f}")
print(f"Max overround: {odds['overround'].max():.4f}")
odds.head()
```
